````python
class Player:
    def __init__(self, symbol, name) -> None:
        self.symbol = symbol
        self.name = name

class Square:
    def __init__(self) -> None:
        self.board = [[1, 2, 3], 
                      [4, 5, 6], 
                      [7, 8, 9]]

class MoveMaker:
    def __init__(self, square) -> None:
        self.position = {
            1: (0, 0), 2: (0, 1), 3: (0, 2),
            4: (1, 0), 5: (1, 1), 6: (1, 2),
            7: (2, 0), 8: (2, 1), 9: (2, 2)
        }
        self.square = square

    def display_board(self):
        for row in self.square.board:
            print(row)

    def make_move(self, spot, player):
        if spot in self.position:
            row, col = self.position[spot]
            if self.square.board[row][col] not in ['X', 'O']:
                self.square.board[row][col] = player.symbol
                print("Great move! Now it's the next player's turn")
                return True 
            else:
                print("That spot is taken, please choose one with a number available")
                return False
        else:
            print("Invalid spot number.")
            return False 

class Game:
    def __init__(self, square):
        self.square = square

    def check_for_winner(self):
        board = self.square.board

        # Check rows
        for row in board:
            if row[0] == row[1] == row[2] and isinstance(row[0], str):
                return f'Winner: {row[0]}'

        # Check columns
        for col in range(3):
            if board[0][col] == board[1][col] == board[2][col] and isinstance(board[0][col], str):
                return f'Winner: {board[0][col]}'

        # Check diagonals
        if board[0][0] == board[1][1] == board[2][2] and isinstance(board[0][0], str):
            return f'Winner: {board[0][0]}'
        if board[0][2] == board[1][1] == board[2][0] and isinstance(board[0][2], str):
            return f'Winner: {board[0][2]}'

        return 'No winner'

````

````python
class GameHistory:
    def __init__(self):
        self.games = {}

    def add_game(self, game_data):
        game_id = str(len(self.games) + 1)
        self.games[game_id] = game_data

    def save_to_file(self, filename='data/game_history.json'):
        with open(filename, 'w') as file:
            json.dump(self.games, file, indent=4)

    def load_from_file(self, filename='data/game_history.json'):
        try:
            with open(filename, 'r') as file:
                self.games = json.load(file)
        except FileNotFoundError:
            print(f"No existing game history found at {filename}. Starting fresh.")
````

````python
from datetime import date

store = {'players': {}, 'games': {}, 'moves': {}, 'game_outcomes': {}}

class Player:
    def __init__(self, name, symbol, four_digits):
        self.name = name
        self.symbol = symbol
        self.last_four_phoneNo = four_digits
        self.id = len(store['players']) + 1
        store['players'][self.id] = {
            'name': self.name, 
            'symbol': self.symbol, 
            'last_four_phoneNo': self.last_four_phoneNo
        }

class Square:
    def __init__(self) -> None:
        self.board = [[1, 2, 3], 
                      [4, 5, 6], 
                      [7, 8, 9]]

class MoveMaker:
    def __init__(self, square) -> None:
        self.position = {
            1: (0, 0), 2: (0, 1), 3: (0, 2),
            4: (1, 0), 5: (1, 1), 6: (1, 2),
            7: (2, 0), 8: (2, 1), 9: (2, 2)
        }
        self.square = square

    def display_board(self):
        for row in self.square.board:
            print(row)

    def make_move(self, spot, player):
        if spot in self.position:
            row, col = self.position[spot]
            if self.square.board[row][col] not in ['X', 'O']:
                self.square.board[row][col] = player.symbol
                print("Great move! Now it's the next player's turn")
                return True 
            else:
                print("That spot is taken, please choose one with a number available")
                return False
        else:
            print("Invalid spot number.")
            return False 

class Game:
    def __init__(self, square):
        self.square = square
        self.id = len(store['games']) + 1
        self.date = date.today().isoformat()
        store['games'][self.id] = {'date': self.date, 'winner_player_id': None}

    def check_for_winner(self):
        board = self.square.board

        # Check rows
        for row in range(3):
            if board[row][0] == board[row][1] == board[row][2] and isinstance(board[row][0], str):
                self.winning_condition = ["row", row + 1]
                return f"Winner: {board[row][0]} \nWinning Condition: Row {self.winning_condition[1]}"

        # Check columns
        for col in range(3):
            if board[0][col] == board[1][col] == board[2][col] and isinstance(board[0][col], str):
                self.winning_condition = ["column", col + 1]
                return f"Winner: {board[0][col]} \nWinning condition: Column {self.winning_condition[1]}"

        # Check diagonals
        if board[0][0] == board[1][1] == board[2][2] and isinstance(board[0][0], str):
            self.winning_condition = ["diagonal", "left"]
            return f"Winner: {board[0][0]} \nWinning Condition: Diagonal Left {self.winning_condition[1]}"
        if board[0][2] == board[1][1] == board[2][0] and isinstance(board[0][2], str):
            self.winning_condition = ["diagonal", "right"]
            return f"Winner: {board[0][2]} \nWinning Condition: Diagonal Right {self.winning_condition[1]}"

        return 'No winner'

    def winning_condition(self):
        return self.winning_condition

class StartGame:
    def __init__(self, player1_name, player2_name):
        self.player1 = Player(player1_name, 'X', 1234)
        self.player2 = Player(player2_name, 'O', 5678)
        # self.player1_moves = []
        # self.player2_moves = []
        self.game = Game(Square())

    def get_player_input(self, player):
        while True:
            try:
                spot = int(input(f"{player.name} ({player.symbol}'s turn). Enter a spot (1-9): "))
                if 1 <= spot <= 9:
                    return spot
                else:
                    print("Please enter a number between 1 and 9.")
            except ValueError:
                print("Please enter a valid number.")

    def start_game(self):
        move_maker = MoveMaker(self.game.square)
        for turn in range(9):
            move_maker.display_board()
            player = self.player1 if turn % 2 == 0 else self.player2
            spot = self.get_player_input(player)
            if move_maker.make_move(spot, player):
                move_record = {'game_id': self.game.id, 'player_id': player.id, 'move': spot}
                store['moves'][len(store['moves']) + 1] = move_record
                # if player == self.player1:
                #     self.player1_moves.append([player.symbol, spot])
                # else:
                #     self.player2_moves.append([player.symbol, spot])

            winner_info = self.game.check_for_winner()
            if winner_info != 'No winner':
                print(winner_info)
                store['games'][self.game.id]['winner_player_id'] = player.id
                store['game_outcomes'][self.game.id] = {'game_id': self.game.id, 
                                                        'winning_condition': self.game.winning_condition}
                break
        else:
            print("The game ended in a draw.")

player1_name = input("Enter name for Player 1 (X): ")
player2_name = input("Enter name for Player 2 (O): ")

square = Square()
move_maker = MoveMaker(square)
game = Game(square)

tictactoe_game = StartGame(player1_name, player2_name)
tictactoe_game.start_game()
````

````python
from datetime import date

store = {'players': {}, 'games': {}, 'moves': {}, 'game_outcomes': {}}

class Player:
    def __init__(self, name, symbol, four_digits):
        self.name = name
        self.symbol = symbol
        self.last_four_phoneNo = four_digits
        self.id = len(store['players']) + 1
        store['players'][self.id] = {
            'name': self.name, 
            'symbol': self.symbol, 
            'last_four_phoneNo': self.last_four_phoneNo
        }

class Square:
    def __init__(self) -> None:
        self.board = [[1, 2, 3], 
                      [4, 5, 6], 
                      [7, 8, 9]]

class MoveMaker:
    def __init__(self, square) -> None:
        self.position = {
            1: (0, 0), 2: (0, 1), 3: (0, 2),
            4: (1, 0), 5: (1, 1), 6: (1, 2),
            7: (2, 0), 8: (2, 1), 9: (2, 2)
        }
        self.square = square

    def display_board(self):
        for row in self.square.board:
            print(row)
    
    def display_board_pp(self):
        return self.square.board

    def make_move(self, spot, player):
        if spot in self.position:
            row, col = self.position[spot]
            if self.square.board[row][col] not in ['X', 'O']:
                self.square.board[row][col] = player.symbol
                print("Great move! Now it's the next player's turn")
                return True 
            else:
                print("That spot is taken, please choose one with a number available")
                return False
        else:
            print("Invalid spot number.")
            return False 

class Game:
    def __init__(self, square):
        self.square = square
        self.date = date.today().isoformat()
        self.winning_condition = None

    def check_for_winner(self):
        board = self.square.board

        # Check rows
        for row in range(3):
            if board[row][0] == board[row][1] == board[row][2] and isinstance(board[row][0], str):
                self.winning_condition = ["row", row + 1]
                return f"Winner: {board[row][0]} \nWinning Condition: Row {self.winning_condition[1]}"

        # Check columns
        for col in range(3):
            if board[0][col] == board[1][col] == board[2][col] and isinstance(board[0][col], str):
                self.winning_condition = ["column", col + 1]
                return f"Winner: {board[0][col]} \nWinning condition: Column {self.winning_condition[1]}"

        # Check diagonals
        if board[0][0] == board[1][1] == board[2][2] and isinstance(board[0][0], str):
            self.winning_condition = ["diagonal", "left"]
            return f"Winner: {board[0][0]} \nWinning Condition: Diagonal Left {self.winning_condition[1]}"
        if board[0][2] == board[1][1] == board[2][0] and isinstance(board[0][2], str):
            self.winning_condition = ["diagonal", "right"]
            return f"Winner: {board[0][2]} \nWinning Condition: Diagonal Right {self.winning_condition[1]}"

        return 'No winner'

    def winning_condition(self):
        return self.winning_condition

class StartGame:
    def __init__(self, player1_name, player2_name):
        self.player1 = Player(player1_name, 'X', 1234)  # Example phone number
        self.player2 = Player(player2_name, 'O', 5678)  # Example phone number
        self.game = Game(Square())
        self.game_id = len(store['games']) + 1
        store['games'][self.game_id] = {'date': self.game.date, 'winner_player_id': None}


    def get_player_input(self, player):
        while True:
            try:
                spot = int(input(f"{player.name} ({player.symbol}'s turn). Enter a spot (1-9): "))
                if 1 <= spot <= 9:
                    return spot
                else:
                    print("Please enter a number between 1 and 9.")
            except ValueError:
                print("Please enter a valid number.")

    def start_game(self):
        move_maker = MoveMaker(self.game.square)
        for turn in range(9):
            move_maker.display_board()
            player = self.player1 if turn % 2 == 0 else self.player2
            spot = self.get_player_input(player)
            if move_maker.make_move(spot, player):
                move_record = {'game_id': self.game_id, 'player_id': player.id, 'symbol': player.symbol, 'position': spot}
                store['moves'][len(store['moves']) + 1] = move_record

            winner_info = self.game.check_for_winner()
            if winner_info != 'No winner':
                print(winner_info)
                store['games'][self.game_id]['winner_player_id'] = player.id
                store['game_outcomes'][self.game_id] = {'game_id': self.game_id, 
                                                        'winning_condition': self.game.winning_condition,
                                                        "final_board_state": move_maker.display_board_pp()}
                break
        else:
            print("The game ended in a draw.")

player1_name = input("Enter name for Player 1 (X): ")
player2_name = input("Enter name for Player 2 (O): ")

square = Square()
move_maker = MoveMaker(square)
game = Game(square)

tictactoe_game = StartGame(player1_name, player2_name)
tictactoe_game.start_game()
````

````python
### Work on sqlite
import sqlite3
import pandas as pd

def db_connection(db_path:str):
    """
    Creates a connection and a cursor to the SQLite database.

    Parameters:
    db_path (str): Path to the SQLite database file.

    Returns:
    tuple: A tuple containing the connection and cursor to the database.
    """
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    return conn, cursor

def load_data_to_sqlite(db_path: str, df, table_name: str) -> None:
    """
    Loads data from a CSV URL into an SQLite table.
    """
    conn, cursor = db_connection(db_path)
    #df = pd.read_csv(data_url)
    df.to_sql(table_name, conn, index=False, if_exists='replace')
    conn.close()
        
def get_table_names(db_path: str) -> list[type]:
    """
    Retrieves a list of all table names in the SQLite database.
    """
    conn, cursor = db_connection(db_path)
    cursor.execute('SELECT name from sqlite_master where type="table"')
    tables = cursor.fetchall()
    conn.close()
    return tables      

def get_table_info(db_path:str, table_name: str) -> list[type]:
    """
    Retrieves the schema information of a given table.
    """
    conn, cursor = db_connection(db_path)
    cursor.execute(f'PRAGMA table_info({table_name})')
    info = cursor.fetchall()
    conn.close()
    return info

def run_query(db_path: str, statement: str) -> list[type]:
    """
    Executes a SQL statement and returns the results.
    """
    conn, cursor = db_connection(db_path)
    cursor.execute(statement)
    results = cursor.fetchall()
    conn.close()
    return results
````

````python

db_path = '../data/TictactoeHistory.db'

# create players
statement = """
CREATE TABLE players (
    id INT PRIMARY KEY,
    name VARCHAR(255) NOT NULL,
    symbol VARCHAR(255),	
    last_four_phoneNo INT	
);
"""
run_query(db_path, statement)

# create game
statement = """
CREATE TABLE games (
    id INT PRIMARY KEY,
    date DATE NOT NULL,
    winner_player_id INT,
    FOREIGN KEY (winner_player_id) REFERENCES players(id)
);
"""
run_query(db_path, statement)

# create moves
statement = """
CREATE TABLE moves (
    id INT PRIMARY KEY,
    game_id INT,
    player_id INT,
    position INT NOT NULL,
    symbol CHAR(1) NOT NULL,
    FOREIGN KEY (game_id) REFERENCES games(id),
    FOREIGN KEY (player_id) REFERENCES players(id)
);
"""

run_query(db_path, statement)

# create moves
statement = """
CREATE TABLE game_outcomes (
    id INT PRIMARY KEY,
    game_id INT,
    winning_condition VARCHAR(255),
    final_board_state VARCHAR(255),
    FOREIGN KEY (game_id) REFERENCES games(id)
);

"""
run_query(db_path, statement)
````
